### Data Preparation For Machine Learning

```python
# Dropping the columns that are not important
columns_to_drop = ["Date","Country","Country ISO","Admin 1",'Event ID', 'Date Event Entered',
       'Date Event Modified','All Food Security Categories']


df.drop(columns_to_drop,axis=1,inplace=True)
```

```python
# i am going to keep only the 4th most frequent target values in my data
# which are aid Systems, Food Production, Food Supply and People Carrying Out Food Activity

top_df = df['Food System Impact'].value_counts().nlargest(4).index
print(top_df)
```

```python
my_df = df.loc[df['Food System Impact'].isin(top_df)] # This is the filtering condition part. The .isin() method checks every single value in the 'Food System Impact' column and returns True if the value is present in the top_df list, and False otherwise. It produces a Boolean Series.
```

```python
# converting the target column into numerical format
my_df["Food System Impact"].unique() # all the rare or irrelevant categories have been successfully removed.
```

```python
# converting the target column categories into numerical format
categorical_mapping = {"Aid Systems":0,"Food Supply":1,"Food Production":2,"People Carrying Out Food Activity":3}

my_df["Food System Impact"] = my_df["Food System Impact"].map(categorical_mapping)  # categorical encoding. Applies the mapping key (categorical_mapping) to each row in the "Food System Impact" column. If a value in the column is found as a key in the mapping, it is replaced by the corresponding value.
```

```python
# making a list of columns to apply label encoder
label_encoder_columns = ['Geo Precision', 'Reported Perpetrator',
       'Reported Perpetrator Name', 'Weapon Carried/Used']


# initiating label encoder
le =  LabelEncoder()


# using for loop to iterate over columns one by one and lable encode them
for i in label_encoder_columns:
    my_df[i] = le.fit_transform(my_df[i])

# iterate through each column name contained in the list label_encoder_columns ('Weapon_Used', then 'Geo_Precision', etc.)

# all the unique labels/text values present in the current column (my_df[i]) and assigns a unique integer to each one ('Low', 'Medium','High'). The encoder applies this mapping, turning the text labels into newly assigned numbers.
```

```python
# checking the data types of columns
my_df.dtypes
```

```python
# as longitude and latitude column are encoded as object data type so changing their data type to float
my_df['Latitude'] = my_df['Latitude'].str.replace(',', '.').astype(float)
my_df['Longitude'] = my_df['Longitude'].str.replace(',', '.').astype(float)
```

```python
my_df.dtypes
```

```python
# making the heatmap to check the correlation among features and target variable
corr = my_df.corr()


sns.heatmap(corr,annot=True,cmap = "coolwarm", fmt=".2f") #rounding
plt.title("Correlation Heat Map")
plt.show()
```

```python
# splitting the data into features and target
X = my_df.drop("Food System Impact",axis=1)
y = my_df[["Food System Impact"]]
```

```python
# Splitting the data into training and testing set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=21)
```

```python
X_train.columns
```

### Machine Learning Classification
- I am using the columns Latitude, Longitude, Geo Precision, Reported Perpetrator,
       Reported Perpetrator Name, Weapon Carried/Used as features to predict the impact on food distribution system.

### Decision Tree Classifier

```python
# initiating the model
dtc_model = DecisionTreeClassifier()

# traiing the model on training data
dtc_model.fit(X_train,y_train)


# getting prediction on testing data
dtc_y_pred = dtc_model.predict(X_test)
```

```python
# calculating accuracy 
dtc_accuracy = metrics.accuracy_score(y_test,dtc_y_pred)

print(f"Accuracy of Decision Tree on Testing data is {dtc_accuracy}")
```

```python
# calculating classification report
dtc_cr =  metrics.classification_report(y_test,dtc_y_pred)
print(dtc_cr)
```

```python
# confusion matrix of decision tree
dtc_cm = metrics.confusion_matrix(y_test,dtc_y_pred)

sns.heatmap(dtc_cm,annot=True, cmap = 'coolwarm', fmt=".2f")
plt.title("Confusion Matrix of DTC")
plt.show
```

### Random Forest Model

```python
# initiating the model
rfc_model = RandomForestClassifier()

# traiing the model on training data
rfc_model.fit(X_train,y_train)


# getting prediction on testing data
rfc_y_pred = rfc_model.predict(X_test)
```

```python
# calculating accuracy 
rfc_accuracy = metrics.accuracy_score(y_test,rfc_y_pred)

print(f"Accuracy of Random Forest on Testing data is {rfc_accuracy}")
```

```python
# calculating classification report
rfc_cr =  metrics.classification_report(y_test,rfc_y_pred)
print(rfc_cr)
```

```python
# confusion matrix of decision tree
rfc_cm = metrics.confusion_matrix(y_test,rfc_y_pred)

sns.heatmap(rfc_cm,annot=True,cmap = "coolwarm", fmt=".2f")
plt.title("Confusion Matrix of RFC")
plt.show()
```
